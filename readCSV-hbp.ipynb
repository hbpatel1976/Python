```python
import csv
grade = []
counter = 0
with open("/content/student.csv") as dataFile:
  data = csv.reader(dataFile)
  for row in data:
    cpi = float(row[2])
    if(cpi>9):
      grade.append("Excellent")
    elif(cpi>8):
      grade.append("Very Good")
    elif(cpi>7):
      grade.append("Good")
    elif(cpi>6):
      grade.append("Average")
    else:
      grade.append("Poor")
    print(f"{row[1]} has '{grade[counter]}' grade'")
    counter+=1
```
